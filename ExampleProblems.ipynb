**Задача 1. Длиннейшая общая подпара**

---



Будем называть пару строк $(α,β)$ подпарой строки γ если $γ=γ_1αγ_2βγ_3$ для некоторых (возможно пустых) строк $γ_1$, $γ_2$ и $γ_3$. Длиной пары строк будет называть сумму длин составляющих ее строк: $|(α,β)|=|α|+|β|$.

По заданным двум строкам $ξ$ и $η$ найдите их длиннейшую общую подпару, то есть такую пару строк $(α,β)$, что она является подпарой как $ξ$, так и $η$, и ее длина максимальная.

**Входные данные**

Входной файл содержит две непустые строки $ξ$ и $η$, состоящие из маленьких букв латинского алфавита. Длина каждой из строк не превышает 3000.

**Выходные данные**

Выведите $α$ на первой строке выходного файла и $β$ на второй строке.

```
s1 = input()
s2 = input()
maxSuffix = [[0 for _ in range(len(s2) + 1)] for _ in range(len(s1) + 1)]
# dp
for i in range(1,len(s1) + 1):
    for j in range(1,len(s2) + 1):
        if s1[i-1] == s2[j-1]:
            maxSuffix[i][j] = maxSuffix[i-1][j-1] + 1
        else:
            maxSuffix[i][j] = 0

maxLen = [[0 for _ in range(len(s2) + 1)] for _ in range(len(s1) + 1)]
for i in range(1,len(s1) + 1):
    for j in range(1,len(s2) + 1):
        maxLen[i][j] = max(maxLen[i-1][j],maxLen[i][j-1],maxSuffix[i][j])
        
#reverse
s1_r = s1[::-1]
s2_r = s2[::-1]
maxSuffix_r = [[0 for _ in range(len(s2) + 1)] for _ in range(len(s1) + 1)]
# dp
for i in range(1,len(s1) + 1):
    for j in range(1,len(s2) + 1):
        if s1_r[i-1] == s2_r[j-1]:
            maxSuffix_r[i][j] = maxSuffix_r[i-1][j-1] + 1
        else:
            maxSuffix_r[i][j] = 0

maxLen_r = [[0 for _ in range(len(s2) + 1)] for _ in range(len(s1) + 1)]
for i in range(1,len(s1) + 1):
    for j in range(1,len(s2) + 1):
        maxLen_r[i][j] = max(maxLen_r[i-1][j],maxLen_r[i][j-1],maxSuffix_r[i][j])

maxi_i = 0
maxi_j = 0
max_value = 0
for i in range(0,len(s1) + 1):
    for j in range(0,len(s2) + 1):
        if maxLen[i][j] + maxLen_r[len(s1) - i][len(s2) - j] > max_value:
            maxi_i = i
            maxi_j = j
            max_value = maxLen[i][j] + maxLen_r[len(s1) - i][len(s2) - j]

answer1 = ''
i = maxi_i
j = maxi_j
# restoring answer
while i != 0 and j != 0:
    if maxLen[i][j] == maxSuffix[i][j]:
        for letter in s1[i - maxSuffix[i][j]:i]:
            answer1 = answer1 + letter
        break
    elif maxLen[i][j] == maxLen[i-1][j]:
        i = i - 1
    elif maxLen[i][j] == maxLen[i][j-1]:
        j = j - 1 

answer2 = ''
i = len(s1) - maxi_i
j = len(s2) - maxi_j
# restoring answer
while i != 0 and j != 0:
    if maxLen_r[i][j] == maxSuffix_r[i][j]:
        for letter in s1_r[i-maxSuffix_r[i][j]:i]:
            answer2 = answer2 + letter
        break
    elif maxLen_r[i][j] == maxLen_r[i-1][j]:
        i = i - 1
    elif maxLen_r[i][j] == maxLen_r[i][j-1]:
        j = j - 1 

print(answer1)
print(answer2[::-1])
```

**Задача 2. Квадратный корень и квадратный квадрат**

---
Найдите такое число x, что $x^2 + \sqrt{x} = C$, с точностью не менее 6 знаков после точки.

**Входные данные**

В единственной строке содержится вещественное число $1.0  ≤  C ≤  10^10$.

**Выходные данные**

Выведите одно число — искомый $x$.

```
import math
c = float(input())
def f(x):
    return x ** 2 + math.sqrt(x)
eps = 10 ** (-7)
l, r = 0, c
m = (l + r)/2
while abs(f(m) - c) > eps:
    m = (l + r)/2
    y = f(m)
    if y > c:
        r = m
    else:
        l = m
print(m)
```
